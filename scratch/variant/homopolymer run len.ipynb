```python
def getHRUNold(ref, alt, seq):
    basesContext = (len(seq) - max(len(ref), len(alt))) / 2
    if len(ref) > len(alt):  
        # deletion
        lastnt = ref[len(alt)]
        type = 'del'
    elif len(ref) < len(alt):
        # insertion
        lastnt = alt[1]
        type = 'ins'
    else:
        # SNV
        lastnt = ref[0]
        type = 'snv'
        
    # check which is longer - trailing HP, or leading
    leadingHRUN = trailingHRUN = 0
    # leading
    for base in seq[basesContext:]:
        if base != alt[0]:
            break
        leadingHRUN += 1
    
    if type != 'del' or leadingHRUN > 1:
        for base in reversed(seq[:basesContext]):
            if base != ref[0]:
                break
            leadingHRUN += 1
    
    # trailing
    for base in seq[basesContext + len(alt):]:
        if base != lastnt:
            break
        trailingHRUN += 1
        
    for base in reversed(seq[:basesContext + len(alt)]):
        if base != lastnt:
            break
        trailingHRUN += 1

    print seq[:basesContext + len(alt)], (leadingHRUN, trailingHRUN)
    return max(leadingHRUN, trailingHRUN)
        
```
