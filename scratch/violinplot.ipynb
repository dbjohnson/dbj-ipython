```python
%matplotlib inline
import pylab as plot
import seaborn as sns
import random
import numpy as np
import pandas as pd
```

```python
n = 73
data = []
labels = []
for i in xrange(n):
    m = random.randint(1, 30)
    data.append(np.random.normal(size=(m,)))
    labels.append(random.choice('asdf'))

positions = range(n/2)
positions.extend(range(n/2 + 5, n + 5))
```

```python
plot.figure(figsize=(24,8), dpi=300)
plt = sns.violinplot(data, names=labels, positions=positions, inner='stick', color=sns.color_palette("muted", 4))
```
