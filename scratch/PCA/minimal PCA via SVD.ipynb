```python
%matplotlib inline
from matplotlib import pylab as plot
import numpy as np
```

```python
def pca(A):
    eig_vals, eig_vecs = np.linalg.eig(np.cov(A.T))    
    return eig_vals, eig_vecs

def pca_svd(A):
    # center to ensure we find vectors that maximize variance, not moment
    n = A - np.mean(A, axis=0)
#     n /= np.std(A, axis=0)
    U, s, Vt = np.linalg.svd(n, full_matrices=False)
    V = Vt.T

    # sort the PCs by descending order of the singular values (i.e. by the
    # proportion of total variance they explain)
    ind = np.argsort(s)[::-1]
    U = U[:, ind]
    s = s[ind]
    V = V[:, ind]
    return (U, s, V)

# generate sample
# np.random.seed(0)
d = 2
n = 200
mu = np.array([2]*d)
cov = np.identity(d)
s1 = 5
s2 = 3
rho = 0.7
cov = np.array([[s1*s1, s1*s2*rho], [s1*s2*rho, s2*s2]])
A = np.random.multivariate_normal(mu, cov, n)
A -= np.mean(A, axis=0)
print A.shape

plot.figure(figsize=(8,8), dpi=100)

plot.subplot(421)
plot.scatter(A[:,0], A[:,1], facecolor='w')
plot.axis([-20, 20, -10, 10])
plot.title('Original Data')


U, s, V = pca_svd(A)

eig_vals, eig_vecs = pca(A)

for pc in ['all', 0, 1]:
    ev = np.copy(eig_vecs)
    S = np.diag(s)

    if pc == 0:
        label = 'first'
        plot.subplot(425)
        ev[:,1:] = 0
        S[1, 1] = 0
    elif pc == 1:
        label = 'second'
        plot.subplot(427)
        ev[:,:1] = 0
        S[0, 0] = 0
    else:
        plot.subplot(423)
        label = 'all'

    projected = A.dot(ev)
    Ahat = ev.dot(projected.T).T
    plot.scatter(Ahat[:,0], Ahat[:,1], facecolor='w')
    plot.axis('equal')
    plot.title('Projected to %s PC - covariance method' % label)
    plot.axis([-20, 20, -10, 10])
    print "Eig - using %s PC, MSE = %.6G" % (label, np.mean((A - Ahat)**2))    

    if pc == 0:
        plot.subplot(426)
    elif pc == 1:
        plot.subplot(428)
    else:
        plot.subplot(424)
        
    Ahat2 = np.dot(U, np.dot(S, V.T))
    Ahat2 += np.mean(A, axis=0)
#     Ahat2 *= np.std(A, axis=0)
    print "SVD - using %s PC, MSE = %.6G" % (label, np.mean((A - Ahat2)**2))    
        
    plot.scatter(Ahat2[:,0], Ahat2[:,1], facecolor='w')
    plot.axis('equal')
    plot.title('Projected to %s PC - SVD' % label)
    plot.axis([-20, 20, -10, 10])
    
plot.tight_layout()
```
