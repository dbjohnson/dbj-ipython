```python
def quantize(values, percentiles=None, floor=None, ceil=None, num_quant_levels=5):
    if percentiles is not None:
        floor, ceil = np.percentile(values.flat, percentiles)

    qtize = np.vectorize(lambda x: int(np.round((num_quant_levels - 1) *
                                       (min(ceil, max(floor, x)) - floor)/(ceil - floor))))
    return qtize(values)


def connected_components(heatmap_quant, neighbors=8):
    zones, num = measure.label(heatmap_quant, neighbors=neighbors, return_num=True, background=-1)
    return zones, num
```

```python
im = diff_images[1][2000:3000, 2000:3000]
nquant = 5
quantized = quantize(im, percentiles=[5, 95], num_quant_levels=nquant)
```

```python
composite = np.zeros(im.shape)
for i in xrange(nquant):
    cc, numzones = connected_components(quantized==i)
    for z in xrange(numzones):
        if np.sum(cc == z) >= 0:
            cc[cc == z] = -1

    composite[cc != -1] = i + 1

plt.subplot(131)
plt.imshow(im)
plt.subplot(132)
plt.imshow(quantized)
plt.subplot(133)
plt.imshow(composite)
```
