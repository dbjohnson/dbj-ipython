```python
import matplotlib
%matplotlib inline
import pylab as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats
import itertools
from IPython.html import widgets
from collections import defaultdict
from collections import OrderedDict

matplotlib.rcParams['figure.figsize'] = (14, 8)
matplotlib.rcParams['figure.dpi'] = 80
```

```python
df_yield = pd.read_csv('C656319A-40A3-3A01-95DE-A09FDBDF48D8.csv', usecols=['Year', 'State', 'Commodity', 'Data Item', 'Value'])
df_yield = df_yield[df_yield['Data Item'] != 'CORN, SILAGE - YIELD, MEASURED IN TONS / ACRE']
df_yield.sort(['Year', 'Commodity', 'State'], inplace=True)
commodity_to_yield_units = pd.pivot_table(df_yield, index='Commodity', values='Data Item', 
               aggfunc=lambda x: set(x).pop().split('MEASURED IN ')[-1].lower().capitalize().replace(' ', '')).to_dict()
```

Load and prepare crop yield data

Print historical yield data for Michigan, US, and whatever local data is available

```python
for i, (commodity, df_c) in enumerate(df_yield.groupby('Commodity')):
    plt.subplot(3,2,i+1)
    plt.hold(True)
    unit = commodity_to_yield_units[commodity]
    df_c.rename(columns={'Value': unit}, inplace=True)
    for location, df_cl in df_c.groupby('State'):
        plt.plot(df_cl['Year'], df_cl[unit], label=location)

    plt.legend(loc=0)
    plt.title('Historical Yields - %s' % commodity.lower().capitalize())
    plt.xlabel('Year')
    plt.ylabel(unit)
    
plt.tight_layout();
```

Load and prepare weather data

```python
# df_weather = pd.read_csv('450698.csv', usecols=['STATION_NAME', 'DATE', 'TMIN', 'TMAX', 'PRCP'])
# df_weather = df_weather[df_weather['STATION_NAME'] == 'GRAND RAPIDS MI US']

df_weather = pd.read_csv('449958.csv', usecols=['STATION_NAME', 'DATE', 'TMIN', 'TMAX', 'PRCP'])
df_weather = df_weather[df_weather['STATION_NAME'] == 'ANN ARBOR UNIVERSITY OF MI MI US']
df_weather['DATE'] = pd.to_datetime(df_weather['DATE'], format='%Y%m%d')
df_weather.sort('DATE', inplace=True)
df_weather['Year'] = df_weather['DATE'].map(lambda x: x.year)
df_weather['Month'] = df_weather['DATE'].map(lambda x: x.month)
df_weather['Day of year'] = df_weather['DATE'].map(lambda x: x.dayofyear)
df_weather.set_index('DATE');

# replace the -9999 values with nans to be more explicit
df_weather = df_weather.applymap(lambda x: np.nan if x == -9999 else x)

df_weather['Max Temp Fahrenheit'] = \
    df_weather['TMAX'].map(
    lambda x: (x / 10.0) * 9.0/5 + 32)

df_weather['Min Temp Fahrenheit'] = \
    df_weather['TMIN'].map(
    lambda x: (x / 10.0) * 9.0/5 + 32)
```

Calculate some annual point metrics that can be correlated with yield

```python
label_to_factor = dict()  # would typically do factor-to-label dict, but dataframes aren't hashable
last_freeze = pd.pivot_table(
                    df_weather[~np.isnan(df_weather['Min Temp Fahrenheit']) &
                               (df_weather['Month'] < 5) & 
                               (df_weather['Min Temp Fahrenheit'] < 32)],
                    index=['Year'], values=['Day of year'], aggfunc=max)
label = 'Date of Last Freeze'
last_freeze.rename(columns={'Day of year': label}, inplace=True)
label_to_factor[label] = last_freeze

avg_max_summer_temp = pd.pivot_table(
                        df_weather[(df_weather['Month'] > 4) & (df_weather['Month'] < 10)],
                        index=['Year'], values=['Max Temp Fahrenheit'], aggfunc=np.mean)
label = 'Avg. Max Summer Temp (F)'
avg_max_summer_temp.rename(columns={'Max Temp Fahrenheit': label}, inplace=True)
label_to_factor[label] = avg_max_summer_temp

# TODO: Should look at precipitation from ~Oct. - ~Sept instead of Jan.-Dec.
total_precip = pd.pivot_table(df_weather, index=['Year'], values=['PRCP'], 
                              aggfunc=lambda x: sum(x)/254.)  # convert 10ths of mm to inches
label = 'Total Annual Precip.'
total_precip.rename(columns={'PRCP': label}, inplace=True)
label_to_factor[label] = total_precip


# TODO: Should look at precipitation from ~Oct. - ~Sept instead of Jan.-Dec.
for month, df_month in df_weather.groupby('Month'):
    month_precip = pd.pivot_table(df_month, index=['Year'], values=['PRCP'], 
                              aggfunc=lambda x: sum(x)/254.)
    months = ['Jan.', 'Feb.', 'March', 'April', 'May', 'June', 'July', 'Aug.', 'Sept.', 'Oct.', 'Nov.', 'Dec.']
    label = '%s Total Precip.' % months[month-1]
    month_precip.rename(columns={'PRCP': label}, inplace=True)
    label_to_factor[label] = month_precip


cmd_to_factor_label_to_pval_df = defaultdict(dict)
for commodity, df_c in df_yield.groupby('Commodity'):
    df_yield_pivot = df_c.pivot(index='Year', columns='State', values='Value')
    df_yield_pivot = df_yield_pivot[~np.isnan(df_yield_pivot['MICHIGAN']) & 
                                    ~np.isnan(df_yield_pivot['US TOTAL'])]
    df_yield_pivot['Normalized Yield'] = df_yield_pivot['MICHIGAN'] / df_yield_pivot['US TOTAL']
    df_yield_pivot.rename(columns={'MICHIGAN': 'Michigan Yield'}, inplace=True)

    for label, df_factor in label_to_factor.items():
        df = pd.merge(df_yield_pivot, df_factor, left_index=True, right_index=True)
        df = df[~np.isnan(df[label]) & ~np.isnan(df['Michigan Yield'])]
        r, p = scipy.stats.pearsonr(df[label], df['Michigan Yield'])
        s = '%s -- p: %s' % (label, ('%1.2e' % p) if p < 0.01 else ('%1.2f' % p))
        cmd_to_factor_label_to_pval_df[commodity][s] = (p, df)
```

```python
class InteractWrapper:
    commodities = sorted(set(df_yield['Commodity']))
    sorted_keys = sorted(cmd_to_factor_label_to_pval_df[commodities[0]].keys(), 
                         key=lambda factor: cmd_to_factor_label_to_pval_df[commodities[0]][factor][0])
    last_commodity = commodities[0]
    factor_widget = widgets.DropdownWidget(values=sorted_keys, value=sorted_keys[0])

    def jointplot_wrapper(self, commodity, factor, kde=False):
        if commodity != self.last_commodity:
            # reset factor widget order and descriptions with appropriate p-values for the selected commodity
            self.last_commodity = commodity
            self.sorted_keys = sorted(cmd_to_factor_label_to_pval_df[commodity].keys(), 
                                 key=lambda factor: cmd_to_factor_label_to_pval_df[commodity][factor][0])

            values = OrderedDict()
            for key in self.sorted_keys:
                values[key] = key
            self.factor_widget.values = values
            self.factor_widget.value = self.sorted_keys[0]
            return


        p, df = cmd_to_factor_label_to_pval_df[commodity][factor]
        label = factor.split(' -- ')[0]
        unit = commodity_to_yield_units[commodity]
        plt.subplot(121)
        plt.plot(df.index, df['Michigan Yield'], label='Michigan Yield')
        plt.plot(df.index, df['US TOTAL'], label='US TOTAL')
        plt.plot(df.index, df[label], label=label)
        plt.legend(loc=0)
        plt.title('Historical Yields - %s' % commodity.lower().capitalize())
        plt.xlabel('Year')
        plt.ylabel(unit)
        plt.subplot(122)
        plt.scatter(df[label], df['Normalized Yield'])
        plt.xlabel(label)
        plt.ylabel('Normalized Yield')
        plt.title('Normalized Yield vs. %s' % label)

        with sns.axes_style("white"):
            sns.jointplot(df[label], df['Michigan Yield'], kind='kde' if kde else 'hex')
            sns.jointplot(df[label], df['Normalized Yield'], kind='kde' if kde else 'hex')


iw = InteractWrapper()
widgets.interact(lambda commodity, factor, kde: iw.jointplot_wrapper(commodity, factor, kde), 
                 commodity=commodities, factor=iw.factor_widget, kde=False);    
```
