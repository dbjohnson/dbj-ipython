```python
%matplotlib inline
import pylab as plt
import seaborn as sns
from IPython.html import widgets
from IPython import display
from spanner import flatfile
from collections import OrderedDict
```

```python
keyfile = '/Users/bryan/projects/bean_counter/data/logs/Trace_JDS680STS21_06.19.14_2.log.src_pgn_spn_key.txt'
# keyfile = '/Users/bryan/projects/bean_counter/data/logs/2013-11-11-JD 7760 cotton picker #3 - operator Adrian - Clark Farm - trace started then machine started - 11.11.13.log.src_pgn_spn_key.txt'
# keyfile = '/Users/bryan/projects/bean_counter/data/logs/Trace_CNH4430Patriot_03.19.14-01.log.src_pgn_spn_key.txt'
source_to_device = flatfile.load_dict(keyfile, ['Source'], ['Device'])
source_to_pgns = flatfile.load_dict(keyfile, ['Source'], ['PGN Name'])
pgn_name_to_id = flatfile.load_dict(keyfile, ['PGN Name'], ['PGN'])
pgn_to_spns = flatfile.load_dict(keyfile, ['PGN'], ['SPN Name'])
pgn_spn_name_to_id = flatfile.load_dict(keyfile, ['PGN', 'SPN Name'], ['SPN'])
spn_to_units = flatfile.load_dict(keyfile, ['SPN'], ['Units'])
```

```python
def sorted_list(vals):
    if not isinstance(vals, list):
        return [vals]
    else:
        return sorted(vals)
    
loaded_traces = dict()
    
src_selector = widgets.DropdownWidget(description='Source',
                                      values=\
                                      OrderedDict(('--'.join([str(src), source_to_device[src]]), src)
                                                     for src in sorted_list(source_to_device.keys())))
pgn_selector = widgets.DropdownWidget(description='PGN')
spn_selector = widgets.DropdownWidget(description='SPN')

def set_pgns():
    pgn_selector.values = OrderedDict((str(pgn_name), pgn_name_to_id[pgn_name])
                                     for pgn_name in sorted_list(pgn_name_to_id.keys())
                                     if pgn_name in source_to_pgns[src_selector.value])
    pgn_selector.value = pgn_selector.values.values()[0]
    
def set_spns():
    pgn = pgn_selector.value

    spn_selector.values = OrderedDict((str(spn_name), pgn_spn_name_to_id[(pgn, spn_name)])
                                     for _pgn, spn_name in pgn_spn_name_to_id
                                     if _pgn == pgn)
    spn_selector.value = spn_selector.values.values()[0]
    
def plot_spn_trace():
    src = src_selector.value
    spn = spn_selector.value
    pgn = pgn_selector.value
    
    
    tracefile = keyfile.replace('src_pgn_spn_key', 'pgn_%d' % pgn)
    if tracefile not in loaded_traces:
        loaded_traces[tracefile] = flatfile.load_dict(tracefile, ['Source', 'SPN'], ['Time', 'Value'])
    src_spn_to_time_val = loaded_traces[tracefile]
    
    display.clear_output(wait=True)
    records = src_spn_to_time_val[(src, spn)]
    if len(records) > 0:
        times, vals = zip(*records)
        plt.figure(figsize=(14,8))
        plt.plot(times, vals, 'b.')
        plt.xlabel('Time')
        plt.ylabel(spn_to_units[spn])

src_selector.on_trait_change(set_pgns, 'value')
pgn_selector.on_trait_change(set_spns, 'value')
spn_selector.on_trait_change(plot_spn_trace, 'value')
display.display(src_selector)
display.display(pgn_selector)
display.display(spn_selector)
set_pgns()
```
