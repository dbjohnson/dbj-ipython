```python
import matplotlib
%matplotlib inline
import pylab as plt
import seaborn as sns
import pandas as pd
import numpy as np
import scipy.stats
import scikits.bootstrap
import itertools
import warnings
warnings.simplefilter('ignore', scikits.bootstrap.InstabilityWarning)

matplotlib.rcParams['figure.figsize'] = (14, 8)
matplotlib.rcParams['figure.dpi'] = 80
```

Load and prepare crop yield data

```python
dfs = []
for csv in ('SUGARBEETS-AcreageYieldProductionandPrice-2014-12-29.csv', 
            'BEANS-AcreageYieldandProductionIrrigatedNonIrrigated-2014-12-29.csv'):
    df = pd.read_csv(csv)
    if csv.startswith('BEANS'):
        df.rename(columns={'YIELD in LB / ACRE': 'YIELD in TONS / ACRE'}, inplace=True)
        
    df = df[['YEAR', 'LOCATION', 'COMMODITY', 'YIELD in TONS / ACRE']]

    def cast_to_float(x):
        try:
            return float(x.strip().replace(',', ''))
        except:
            return np.nan

    df['YIELD in TONS / ACRE'] = df['YIELD in TONS / ACRE'].map(cast_to_float)
    df = df[~np.isnan(df['YIELD in TONS / ACRE'].tolist())]
    if csv.startswith('BEANS'):
        df['YIELD in TONS / ACRE'] = df['YIELD in TONS / ACRE'].map(lambda x: x / 2000.)
    
    df = df[(df['COMMODITY'] == 'BEANS, DRY EDIBLE') |
            (df['COMMODITY'] == 'SUGARBEETS')]
#     df = df[(df['LOCATION'] == 'MICHIGAN') |
#             (df['LOCATION'] == 'US TOTAL')]
    
    dfs.append(df)

df_yield = pd.concat(dfs)
df_yield.sort(['YEAR', 'COMMODITY', 'LOCATION'], inplace=True)
# pd.pivot_table(df_yield, index=['COMMODITY', 'LOCATION'], aggfunc=lambda x: sum([1 if xx != np.nan else 0 for xx in x]))
```

Print historical yield data for Michigan, US, and whatever local data is available

```python
for i, (commodity, df_c) in enumerate(df_yield.groupby('COMMODITY')):
    plt.subplot(2,1,i+1)
    plt.hold(True)

    reg_labeled = False
    for location, df_cl in df_c.groupby('LOCATION'):
        if location not in ('MICHIGAN', 'US TOTAL'):
            plt.plot(df_cl['YEAR'], df_cl['YIELD in TONS / ACRE'], 'r.', alpha=0.2, 
                     label='Local' if not reg_labeled else None)
            reg_labeled = True
        else:
            plt.plot(df_cl['YEAR'], df_cl['YIELD in TONS / ACRE'], label=location)

    plt.legend()
    plt.title('Historical Yields - %s' % commodity.lower().capitalize())
    plt.xlabel('Year')
    plt.ylabel('Tons / Acre')
plt.tight_layout();
```

Load and prepare weather data

```python
df_weather = pd.read_csv('449958.csv')
df_weather = df_weather[df_weather['STATION_NAME'] == 'ANN ARBOR UNIVERSITY OF MI MI US']

df_weather.rename(columns={'WT05': 'Hail (may include small hail)',
'WT06': 'Glaze or rime',
'WT03': 'Thunder',
'WT04': 'Ice pellets, sleet, snow pellets, or small hail',
'SNOW': 'Snowfall (mm)',
'WT07': 'Dust, volcanic ash, blowing dust, blowing sand, or blowing obstruction',
'WT08': 'Smoke or haze',
'WESD': 'Water equivalent of snow on the ground (tenths of mm)',
'TMAX': 'Maximum temperature (tenths of degrees C)',
'SNWD': 'Snow depth (mm)',
'SN52': 'Minimum soil temperature (tenths of degrees C) with sod cover at 10 cm depth',
'TOBS': 'Temperature at the time of observation (tenths of degrees C)',
'WT11': 'High or damaging winds',
'WT10': 'Tornado, waterspout, or funnel cloud',
'SN32': 'Minimum soil temperature (tenths of degrees C) with bare ground cover at 10 cm depth',
'WT14': 'Drizzle',
'WT16': 'Rain (may include freezing rain, drizzle, and freezing drizzle)',
'TMIN': 'Minimum temperature (tenths of degrees C)',
'WESF': 'Water equivalent of snowfall (tenths of mm)',
'PRCP': 'Precipitation (tenths of mm)',
'SX52': 'Maximum soil temperature (tenths of degrees C) with sod cover at 10 cm depth',
'TSUN': 'Daily total sunshine (minutes)',
'SX32': 'Maximum soil temperature (tenths of degrees C) with bare ground cover at 10 cm depth',
'WT02': 'Heavy fog or heaving freezing fog (not always distinguished from fog)',
'WT01': 'Fog, ice fog, or freezing fog (may include heavy fog)'}, inplace=True)
df_weather['DATE'] = pd.to_datetime(df_weather['DATE'], format='%Y%m%d')
df_weather.sort('DATE', inplace=True)
df_weather['year'] = df_weather['DATE'].map(lambda x: x.year)
df_weather['month'] = df_weather['DATE'].map(lambda x: x.month)
df_weather['day'] = df_weather['DATE'].map(lambda x: x.day)
df_weather['day of year'] = df_weather['DATE'].map(lambda x: x.dayofyear)
df_weather = df_weather.applymap(lambda x: np.nan if x == -9999 else x)
df_weather.set_index('DATE');

df_weather['Max Temp Fahrenheit'] = \
    df_weather['Maximum temperature (tenths of degrees C)'].map(
    lambda x: (x / 10.0) * 9.0/5 + 32)

df_weather['Min Temp Fahrenheit'] = \
    df_weather['Minimum temperature (tenths of degrees C)'].map(
    lambda x: (x / 10.0) * 9.0/5 + 32)
```

Calculate some annual point metrics that can be correlated with yield

```python
last_freeze = pd.pivot_table(
                    df_weather[~np.isnan(df_weather['Min Temp Fahrenheit']) &
                               (df_weather['month'] < 5) & 
                               (df_weather['Min Temp Fahrenheit'] < 30)],
                    index=['year'], values=['day of year'], aggfunc=max)

last_freeze.rename(columns={'day of year': 'Date of Last Freeze'}, inplace=True)

total_precip = pd.pivot_table(df_weather, index=['year'], values=['Precipitation (tenths of mm)'], 
                              aggfunc=lambda x: sum(x)/254.)
total_precip.rename(columns={'Precipitation (tenths of mm)': 'Annual Precipitation (inches)'}, inplace=True)

avg_max_summer_temp = pd.pivot_table(
                        df_weather[(df_weather['month'] > 4) & (df_weather['month'] < 10)],
                        index=['year'], values=['Max Temp Fahrenheit'], aggfunc=np.mean)
avg_max_summer_temp.rename(columns={'Max Temp Fahrenheit': 'Avg Summer Max Temp'}, inplace=True)

factors = ('Annual Precipitation (inches)', 'Avg Summer Max Temp', 'Date of Last Freeze')
rows = 2 * 2
f, axes = plt.subplots(4, len(factors))
row = 0
for commodity, df_c in df_yield.groupby('COMMODITY'):
    df_yield_pivot = df_c.pivot(index='YEAR', columns='LOCATION', values='YIELD in TONS / ACRE')
    df_yield_pivot = df_yield_pivot[~np.isnan(df_yield_pivot['MICHIGAN']) & 
                                    ~np.isnan(df_yield_pivot['US TOTAL'])]
    normalized_col = '%s\nNormalized Yield' % commodity.lower().capitalize()
    df_yield_pivot[normalized_col] = df_yield_pivot['MICHIGAN'] / df_yield_pivot['US TOTAL']
    michigan_col = '%s\nMichigan Yield' % commodity.lower().capitalize()
    df_yield_pivot.rename(columns={'MICHIGAN':michigan_col}, inplace=True)

    dfnew = pd.merge(df_yield_pivot, last_freeze, left_index=True, right_index=True)
    dfnew = pd.merge(dfnew, total_precip, left_index=True, right_index=True)
    dfnew = pd.merge(dfnew, avg_max_summer_temp, left_index=True, right_index=True)

    # remove any nans
    val_cols = [normalized_col, michigan_col]
    val_cols.extend(factors)
    for col in val_cols:
        dfnew = dfnew[~np.isnan(dfnew[col])]
        

    for y in [normalized_col, michigan_col]:
        for col, x in enumerate(factors):
            r, p = scipy.stats.pearsonr(dfnew[x], dfnew[y])
            ax = axes[row][col]
            with sns.axes_style("white"):
                sns.kdeplot(dfnew[x], dfnew[y], shade=True, ax=ax)
            if row != 3:
                ax.set_xlabel('')
            if col != 0:
                ax.set_ylabel('')
                
#         ax = plt.subplot(2, 4, row*4 + cnt + 1)
#         plt.hexbin(dfnew[x], dfnew[y], axes=ax, gridsize=10, marginals=True)
            sns.jointplot(dfnew[x], dfnew[y], kind='hex')

            def tostr(x):
                if abs(x) < 0.01:
                    return '%1.2e' % x
                else:
                    return '%1.2f' % x

            ax.text(0.05, 0.95, 'pearson_r: %s\np: %s' % (tostr(r), tostr(p)),
                    ha='left', va='top', transform=ax.transAxes)
        row += 1

plt.tight_layout()
```

Temps by decade

```python
def tsplot_sparse(df, time, value, condition, ci):
    plt.hold(True)
    for cc, df_condition in df.groupby(condition):
        t = []
        high = []
        low = []
        central = []
        subj = []
        for tt, df_time_condition in df_condition.groupby(time):
            df_time_condition.sort(time, inplace=True)
            tcvals = df_time_condition[value]
            if ci == 1:
                interval = (min(tcvals), max(tcvals))
            else:
                interval = scikits.bootstrap.ci(tcvals, alpha=(1.0-ci)/2, n_samples=100)
#             ci(data, statfunction=np.average, alpha=0.05, n_samples=10000, method='bca', output='lowhigh', epsilon=0.001, multi=None)
            t.append(tt)
            low.append(interval[0])
            high.append(interval[1])
            central.append(np.mean(tcvals))
        h, = plt.plot_date(t, central, label=cc, fmt='-')
        plt.gca().fill_between(t, low, high, alpha=0.2, color=h.get_color())

    plt.legend(loc=0)
    
df_weather['decade'] = df_weather['year'].map(lambda x: '%ds' % (x - (x % 10)))    
tsplot_sparse(df_weather, 'day of year', 'Max Temp Fahrenheit', condition='decade', ci=1)
```
