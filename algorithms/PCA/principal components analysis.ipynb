```python
%matplotlib inline
from matplotlib import pylab as plot
import numpy as np
```

```python
# generate sample
# np.random.seed(10)
d = 2
n = 200
mu = np.array([0]*d)
cov = np.identity(d)
s1 = 10
s2 = 3
rho = 0.5
cov = np.array([[s1*s1, s1*s2*rho], [s1*s2*rho, s2*s2]])
s = np.random.multivariate_normal(mu, cov, n)

# covariance/scatter matrix
sc = s - np.mean(s, axis=0)
sc2 = [ss.reshape(-1,1).dot(ss.reshape(1,-1)) for ss in sc]
sm = np.sum(sc2, axis=0)

# check against built-in covariance function
np.testing.assert_array_almost_equal(sm/(n-1), np.cov(s.T), decimal=6, err_msg='', verbose=True)


plot.plot(s[:,0], s[:,1], 'b.')
plot.axis('equal')
plot.title('Original Data');
```

```python
# convenience/wrapper function for np.linalg.eig.  I don't like how it returns the eigenvectors as column vecs
def eigenVectors(m, doPrint=False):
    mags, vecs = np.linalg.eig(np.cov(m.T))
    vecs = vecs.T # transform to row vectors for convenient pairing
    if doPrint:
        for mag, vec in zip(mags, vecs):
            print 'magnitude: %1.2e, vector: %s, norm: %1.2e' % (mag, str(vec), np.linalg.norm(vec))

    return {'magnitudes':mags, 'vectors':vecs}
```

```python
# plot the eigenvectors 
plot.plot(sc[:,0], sc[:,1], 'b.')
plot.hold(True)

eig = eigenVectors(sc, doPrint=True)
for mag, vec in zip(eig['magnitudes'], eig['vectors']):
    plot.axes().arrow(0, 0, vec[0]*mag, vec[1]*mag, width=.1, head_width=0.4, head_length=0.5, fc='r', ec='r')
    
# covariance ellipse - just for fun
theta = np.linspace(0, 2*np.pi, 100)
circle = np.array([np.sin(theta), np.cos(theta)]).T
oneSigma = np.array([np.sqrt(mag)*vec for mag, vec in zip(eig['magnitudes'], eig['vectors'])])
oneSigma = circle.dot(oneSigma)
plot.plot(oneSigma[:,0], oneSigma[:,1], 'r', linewidth=2)
    
plot.axis('equal')
plot.title('Data with Principal Components');
```

```python
# project data onto principal components
w = eig['vectors']
proj = w.dot(sc.T).T
plot.plot(proj[:,0], proj[:,1], 'b.');
plot.axis('equal')
plot.hold(True)

# eigen vectors for projected data should align with x/y
eigProj = eigenVectors(proj, doPrint=True)
for mag, vec in zip(eigProj['magnitudes'], eigProj['vectors']):
    plot.axes().arrow(0, 0, vec[0]*mag, vec[1]*mag, width=.1, head_width=0.4, head_length=0.5, fc='r', ec='r')

plot.title('Projected data and Principal Components');
```
