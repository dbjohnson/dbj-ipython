```python
%matplotlib inline
from matplotlib import pylab as plot
import numpy as np
```

```python
def pca(A):
    M = (A-np.mean(A.T,axis=1)).T # subtract the mean (along columns)
    eig_vals, eig_vecs = np.linalg.eig(np.cov(M))
    return eig_vals, eig_vecs

# generate sample
d = 2
n = 200
mu = np.array([2]*d)
cov = np.identity(d)
s1 = 5
s2 = 3
rho = 0.9
cov = np.array([[s1*s1, s1*s2*rho], [s1*s2*rho, s2*s2]])
A = np.random.multivariate_normal(mu, cov, n)

plot.subplot(221)
plot.scatter(A[:,0], A[:,1], facecolor='w')
plot.axis('equal')

eig_vals, eig_vecs = pca(A)
projected = A.dot(eig_vecs)
plot.subplot(222)
plot.scatter(projected[:,0], projected[:,1], facecolor='w')
plot.axis('equal')

```
