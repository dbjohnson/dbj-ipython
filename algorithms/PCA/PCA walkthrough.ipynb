```python
%matplotlib inline
from matplotlib import pylab as plt
from IPython.html.widgets import interact, fixed
from IPython.html import widgets
from IPython.display import display
import seaborn
import numpy as np
import random
```

```python
class SimDataGenerator(object):
    def __init__(self, num_genes, num_amplicons_per_gene, mu_amp_independent, 
                 std_amp_independent, mu_rho_systematic, std_rho_systematic):
        self.num_genes = num_genes = num_genes
        self.num_amplicons_per_gene = num_amplicons_per_gene
        
        n = self.num_genes * self.num_amplicons_per_gene
        if std_amp_independent > 0:
            sigmas = np.random.normal(loc=mu_amp_independent, scale=std_amp_independent, size=n)
        else:
            sigmas = [mu_amp_independent] * n

        sigmas2 = np.power(sigmas, 2)
        self.cov = np.diag(sigmas2)

        correlated_idx = range(n)
        random.shuffle(correlated_idx)
        
        # make x groups of correlated amplicons
        idx = 0
        for x in xrange(4):
            m = random.randint(5, 10)
            correlated_idx = correlated_idx[idx:idx+m]
            idx += m

            rho = mu_rho_systematic
            for i in correlated_idx:
                for j in correlated_idx:
                    if j <= i:
                        continue

                    s_i = sigmas[i]
                    s_j = sigmas[j]
                    self.cov[i, j] = self.cov[j, i] = rho * s_i * s_j

            
    def draw_normals(self, num_draws):
        n = self.num_genes * self.num_amplicons_per_gene
        # log2(copy number ratio) - drawn at mu=0, covariance
        l2cnr = np.random.multivariate_normal([0] * n, self.cov, num_draws).T
        l2cnr -= np.median(l2cnr, axis=0)
#         l2cnr = np.log2(600*np.power(2, l2cnr+1))
        return l2cnr
    
    def draw(self, num_draws=1):
        l2cnr = self.draw_normals(num_draws)
        # sprinkle in some CNVs
        cn_truth = []
        for gene in xrange(self.num_genes):
            for draw in xrange(num_draws):
                if random.random() < 0.05:
                    if False and random.random() < 0.3:
                        # deletion
                        gene_l2cnr = random.randint(-4, -1) 
                    else:
                        # amplification
                        gene_l2cnr = random.randint(1, 3)  
                    pos = gene*self.num_amplicons_per_gene

                    l2cnr[pos:pos + self.num_amplicons_per_gene, draw] += gene_l2cnr
                    cn_truth.append(2*(2**gene_l2cnr))
                else:
                    cn_truth.append(2)
                    
        return (cn_truth, l2cnr)
    
```

```python
class PCA(object):
    def __init__(self, num_train, num_normals, sdg):
        truth, l2cnr = sdg.draw(num_train)
        diffs = PCA.diffs(l2cnr)
        self.means = np.mean(diffs, axis=1)
        self.stds = np.std(diffs, axis=1)
        
        (U, s, Vt) = np.linalg.svd(self.whiten(diffs)) 
        ind = np.argsort(s)[::-1]
        self.s = s[ind]
        self.U = U[:, ind]
        self.Vt = Vt[ind,:]
    
        normals = sdg.draw_normals(num_normals)
        
        self.npc_to_normal_residuals = dict()
        for npc in xrange(100):
            corrected_normals = self.remove_pcs(normals, npc, remove_normals=False)
            self.npc_to_normal_residuals[npc] = np.median(corrected_normals, axis=1)
        
    
    @staticmethod
    def diffs(vals):
        diffs = np.matrix(np.zeros((vals.shape[0]+1, vals.shape[1])))
        diffs[1:,:] = vals
        return np.diff(diffs, axis=0)

    def whiten(self, d):
        return (d - self.means) / self.stds
        
    def remove_pcs(self, l2cnr, num_pcs_to_remove, remove_normals=True):
        #   xx0 <- (input.v - amean) / asd
        #   pcw <- t(r$u[,ind]) %*% matrix(xx0)
        #   xx1 <- xx0 - r$u[,ind] %*% pcw
        #   xx1 * asd + amean
        xx0 = self.whiten(PCA.diffs(l2cnr))
        pcw = self.U[:, :num_pcs_to_remove].T.dot(xx0)
        xx1 = xx0 - self.U[:, :num_pcs_to_remove].dot(pcw)
        corrected_diffs = np.multiply(xx1, self.stds) + self.means
        corrected = np.cumsum(corrected_diffs, axis=0)

        if remove_normals:
            corrected -= self.npc_to_normal_residuals[num_pcs_to_remove]

        return corrected
```

```python
def sim_wrapper(num_genes=10, num_amplicons_per_gene=10, num_draws=100,
                noise_scale=0.5, systematic_effect=0.9,
                num_pcs_to_remove=4, remove_normal_residuals=True):

    mu_amp_independent = noise_scale
    std_amp_independent = mu_amp_independent/4
    mu_rho_systematic = systematic_effect
    std_rho_systematic = 0 #mu_rho_systematic/4
    
    simulate_and_plot(num_genes=num_genes, 
                    num_amplicons_per_gene=num_amplicons_per_gene,
                    mu_amp_independent=mu_amp_independent, 
                    std_amp_independent=std_amp_independent, 
                    mu_rho_systematic=mu_rho_systematic, 
                    std_rho_systematic=std_rho_systematic,
                    num_pcs_to_remove=num_pcs_to_remove, 
                    num_draws=num_draws, remove_normals=remove_normals)

def simulate_and_plot(num_genes=3, num_amplicons_per_gene=20,
                    mu_amp_independent=0.3, std_amp_independent=0.1, 
                    mu_rho_systematic=1, std_rho_systematic=0,
                    num_pcs_to_remove=1, transpose=0, num_draws=50,
                    remove_normals=False):
    
    # set seed so we get the same result as the sliders are moved
    np.random.seed(0)
    random.seed(0)
    
    global sdg, pca
    sdg = SimDataGenerator(num_genes, num_amplicons_per_gene, mu_amp_independent, 
                           std_amp_independent, mu_rho_systematic, std_rho_systematic)
    
    pca = PCA(num_train=num_draws, num_normals=10, sdg=sdg)
    
    def plot(yvalues, ylabel, subplot, title=None):
        ax = fig.add_subplot(subplot)
        x = np.matrix(xrange(yvalues.shape[0])).T
        x = np.tile(x, (1, yvalues.shape[1]))
        plt.plot(x, yvalues, '.-', alpha=0.8)
        plt.ylabel(ylabel)
        xticks = [i*num_amplicons_per_gene for i in xrange(num_genes)]
        xticklabels = ['gene %d' % (i + 1) for i in xrange(num_genes)]  
        ax.set_xticks(xticks)
        ax.set_xticklabels(xticklabels)    
        if title is not None:
            plt.title(title)
    

    cn_truth, l2cnr = sdg.draw(num_draws)
    diffs = pca.diffs(l2cnr)        
    mapds = np.median(np.abs(np.squeeze(np.asarray(diffs))), axis=0)
    mapd = np.median(mapds)
    fig = plt.figure(figsize=(16, 16), dpi=100)
    plot(l2cnr, 'log2(copy number ratio)', 321, 'Median MAPD: %1.2f' % mapd)
    plot(diffs, 'diff(log2(copy number ratio))', 323)
    
    
#     scaled_pcs = pca.U*np.diag(pca.s)
#     plot(np.abs(scaled_pcs), 'Principal Components', 325)        
    plot(pca.npc_to_normal_residuals[num_pcs_to_remove], 'Normal Residuals', 325, 'Normal Residuals')
    

    corrected = pca.remove_pcs(l2cnr, num_pcs_to_remove, remove_normals)
    corrected_diffs = pca.diffs(corrected)
    
    mapds = np.median(np.abs(np.squeeze(np.asarray(corrected_diffs))), axis=0)
    mapd = np.median(mapds)

    
    plot(corrected, 'corrected', 322, 'Median MAPD: %1.2f' % mapd)
    plot(corrected_diffs, 'corrected diffs', 324)

    raw_cn = []
    corrected_cn = []
    for i in xrange(num_genes):
        for draw in xrange(num_draws):
            pos = i*num_amplicons_per_gene
            l2cnr_raw = np.median(l2cnr[pos:pos + num_amplicons_per_gene, draw])
            l2cnr_corr = np.median(np.array(corrected[pos:pos + num_amplicons_per_gene, draw]))
            raw_cn.append(2**(l2cnr_raw+1))
            corrected_cn.append(2**(l2cnr_corr+1))
        
    plt.subplot(326)
    plt.scatter(cn_truth, raw_cn, c='g')
    plt.hold(True)
    plt.scatter(cn_truth, corrected_cn, c='b')
    mn = min(min(raw_cn), min(raw_cn))
    mx = max(max(raw_cn), max(corrected_cn))
    plt.plot([mn, mx], [mn, mx], 'r')
    plt.xlabel('Truth CN')
    plt.ylabel('Corrected CN')
        

dmy = interact(sim_wrapper, num_genes=widgets.DropdownWidget(values={'3':3, '10':10, '100':100}, value=10),
              num_amplicons_per_gene=widgets.DropdownWidget(values={'3':3, '10':10, '100':100}, value=10), 
              num_draws=widgets.DropdownWidget(values={'10':10, '100':100, '500':500, '1000':1000}, value=100),
              noise_scale=(0.0, 1.0), systematic_effect=(0.0, 1.0),
              num_pcs_to_remove=(0, 100), remove_normal_residuals=False)
```

```python
def draw_one(num_pcs_to_remove=4, remove_normal_residuals=True):
    truth, l2cnr = sdg.draw()
    corrected = pca.remove_pcs(l2cnr, num_pcs_to_remove, remove_normal_residuals)
    
    xx = []
    for cn in truth:
        xx.extend([np.log2(cn/2)]*sdg.num_amplicons_per_gene)
    
    plt.close()
    plt.step(xrange(len(xx)), xx)
    plt.hold(True)
#     plt.step(xrange(len(l2cnr)), np.power(2, l2cnr+1))
#     plt.step(xrange(len(corrected)), np.power(2, corrected+1))
    plt.step(xrange(len(l2cnr)), l2cnr)
    plt.step(xrange(len(corrected)), corrected)
    xticks = [i*sdg.num_amplicons_per_gene for i in xrange(sdg.num_genes)]
    xticklabels = ['gene %d' % (i + 1) for i in xrange(sdg.num_genes)]  
    plt.gca().set_xticks(xticks)
    plt.gca().set_xticklabels(xticklabels)    
    plt.legend(['Truth', 
           'Raw MAPD: %1.2f' % np.median(np.abs(np.diff(l2cnr[:,0]))),
           'Corrected MAPD: %1.2f' % np.median(np.abs(np.diff(np.squeeze(np.asarray(corrected)))))], loc=0
           )
    

interact(draw_one, num_pcs_to_remove=(0,100), remove_normal_residuals=False);
```
