```python
%matplotlib inline
import pylab as plt
import numpy as np
import seaborn as sns
import random
```

```python
grid_size = 5

board = np.zeros((grid_size, grid_size))
for row in xrange(grid_size):
    for col in xrange(grid_size):
        board[row, col] = (row + col) % grid_size + 1

cols_shuffle = range(grid_size)
np.random.shuffle(cols_shuffle)
rows_shuffle = range(grid_size)
np.random.shuffle(rows_shuffle)
board = board[rows_shuffle, :]
board = board[:, cols_shuffle]

plt.figure(figsize=(16,8))
plt.subplot(121)
sns.heatmap(board, annot=True, cmap=sns.cubehelix_palette(grid_size, start=1, rot=2, as_cmap=True))
plt.title('Answer Key')

cage_grid = np.zeros((grid_size, grid_size))
cage_idx = 1

def random_walk():
    return np.random.randint(-1,2)

while np.any(cage_grid) == 0:
    for row in xrange(grid_size):
        for col in xrange(grid_size):
            if cage_grid[row, col] == 0: 
                rr = row
                cc = col
                while rr >= 0 and rr < grid_size and \
                    cc >=0 and cc < grid_size and \
                    cage_grid[rr, cc] == 0:
                    cage_grid[rr, cc] = cage_idx
                    step = random_walk()
                    if step != 0:
                        rr += step
                    else:
                        cc += random_walk()
                        
                cage_idx += 1

plt.subplot(122)
# cmap = sns.dark_palette("palegreen", as_cmap=True)
sns.heatmap(cage_grid, annot=False, cmap=sns.cubehelix_palette(grid_size, start=1, rot=2, as_cmap=True))
plt.title('Cage index');
```
